Parte 2: Dando funcionalidades
Vamos a “darle vida” a nuestro tanque. 

Vamos a completar los métodos con operaciones comunes

En Tanque:
calcularNivel(self):
Debe tomar el volumen actual del tanque, y en base a eso calcular el nivel actual en metros.

Para hacer la conversión, debemos: 

Calcular el volumen total del tanque. Volumen de un cilindro: altura * pi * (diametro/2)^2
Dividir el volumen medido del tanque en litros (se calcula en el método update), por el volumen calculado en el punto anterior
Multiplicar por la altura del tanque para obtener el nivel en metros.
update(self, tiempo – default 1):
Este método calcula el volumen actual de llenado del tanque. 

Sumar los caudales de las válvulas (valvula[numerovalvula].caudalActual) para obtener el caudal resultante.
Al caudal resultante, multiplicarlo por el tiempo de actualización, para obtener cuanta agua entró desde la última llamada. 
Sumar el agua que entró al volumen actual del tanque. 
Limitar el volumen de agua del tanque a los niveles máximos y mínimos, para evitar que se llene o se vacíe demasiado
cargarTanque(self) y vaciarTanque(self):
Itera entre todas las válvulas, y dependiendo del tipo, ejecuta el método “abrirValvula” o “cerrarValvula”

cargarTanque cierra todas las válvulas tipo S, y abre todas las válvulas tipo E
vaciarTanque cierra todas las válvulas tipo E, y abre todas las válvulas tipo S
En Valvula:
abrirValvula y cerrarValvula:
abrirValvula asigna el caudal configurado de la valvula (caudal) a CaudalActual con algunas consideraciones: Si el tipo es “E”, el caudal es positivo, y si es tipo “S” el caudal es negativo. 
cerrarValvula pone CaudalActual a 0

```python
from math import pi

class Tanque:
    def __init__(self, valvulas,diametro=1.0, altura=1.0):
        self.diametro = diametro
        self.altura = altura
        self.valvulas = valvulas

        self.nivelActual = 0
        self.caudal = 0
        self.volumenActual = 0
        

    def calcularNivel(self):
        volumenMaximo = self.altura * pi * (self.diametro/2) ** 2
        self.nivelActual = (self.volumenActual / volumenMaximo) * self.altura
 

    def update(self, tiempo):
        volumen_maximo = pi * (self.diametro / 2) ** 2 * self.altura
        sumaCaudal = sum(valvula.caudalActual for valvula in self.valvulas)
        volumenAdicional = sumaCaudal * tiempo

        self.volumenActual = self.volumenActual + volumenAdicional
        self.volumenActual = max(0, min(self.volumenActual, volumen_maximo))
        pass

    def cargarTanque(self):
        for valvula in self.valvulas:
            if valvula.tipo == "E":
                valvula.abrirValvula()
            elif valvula.tipo == "S":
                valvula.cerrarValvula()
        

    def vaciarTanque(self):
        for valvula in self.valvulas:
            if valvula.tipo == "S":
                valvula.abrirValvula()
            elif valvula.tipo == "E":
                valvula.cerrarValvula()
        

class valvula:
    def __init__(self, tipo = "E", caudal = 1.0):
        self.tipo = tipo
        self.caudal = caudal

        self.caudalActual = 0

    def abrirValbula(self):
        if self.tipo == "E":
            self.caudalActual = self.caudal 
        if self.tipo == "F":
            self.caudalActual = self.caudal * -1


    def cerrarValvula(self):
        self.caudalActual = 0
```
